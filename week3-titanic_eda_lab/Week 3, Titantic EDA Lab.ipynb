```python
df_filled_missing = df.copy()
df_filled_missing.Age = df_filled_missing.Age.fillna(value=df.Age.mean())
df_filled_missing[df_filled_missing.Age.isnull()]
```

```python
#Ages
## Provided data
df.Age.hist(bins=16)
```

```python
## Estimated data
### The spike in the 25-30 yrs. value is because all missing ages were replaced with the average age.
df_filled_missing.Age.hist(bins=16)
```
