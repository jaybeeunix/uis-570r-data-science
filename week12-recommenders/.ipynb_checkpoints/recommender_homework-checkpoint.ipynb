```python
M = pd.read_csv('Movie Reviews - Sheet1.csv', index_col='Name')
```

```python
def pearson(s1, s2):
    """Take two pd.Series objects and return a pearson correlation."""
    s1_c = s1 - s1.mean()
    s2_c = s2 - s2.mean()
    return np.sum(s1_c * s2_c) / np.sqrt(np.sum(s1_c ** 2) * np.sum(s2_c ** 2))
```

```python
def get_recs(movie_name, M, num):

    import numpy as np
    reviews = []
    for title in M.columns:
        if title == movie_name:
            continue
        cor = pearson(M[movie_name], M[title])
        if np.isnan(cor):
            continue
        else:
            reviews.append((title, cor))
    
    reviews.sort(key=lambda tup: tup[1], reverse=True)
    return reviews[:num]

    
```

```python
def zero_unseen(M, unseen):
    for col in M.columns:
        for unsn in unseen:
            #print('chucking {}, {}, which is {}'.format(unsn, col, M.at[unsn, col]))
            M.at[unsn, col] = 0
            
zero_unseen(cooccurance, havent_seen)
```

```python
wts = cooccurance.sum()
wts
```

```python
def normalize(recs, wts):
    new_recs = recs.copy()
    for title in wts.axes[0].tolist():
        new_recs.at[title] = recs.at[title] / wts.at[title]
    return new_recs

new_recs = normalize(recs, wts)
# rm the movies i've already seen
new_recs.drop(labels=have_seen)
```
